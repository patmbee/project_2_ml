```python
# Import the required libraries and dependencies
# Import Modules
import pandas as pd
import numpy as np
import hvplot.pandas
import matplotlib.pyplot as plt
import datetime
from pandas_datareader import data as pdr
import yfinance as yf
from pandas.tseries.offsets import DateOffset
from pathlib import Path
```

```python
# set csv path - metrics csc saved from qqq_backtest notebook
metrics_csv_path = Path('metrics.csv')
```

```python
# read in csv file to df
qqq_df = pd.read_csv(metrics_csv_path,index_col='Exit Date', infer_datetime_format=True, parse_dates=True)
display(qqq_df.head())
display(qqq_df.tail())   
```

```python
# Drop unnammed column
qqq_df.drop(qqq_df.columns[qqq_df.columns.str.contains('Unnamed')],axis=1,inplace=True)
qqq_df.tail()
```

```python
# Create df of entry exit totals for plotting
mom_df=qqq_df.loc[:,['Profit/Loss']]
mom_df.tail()     
```

```python
# Group by year
gb_yoy = mom_df.groupby([(mom_df.index.year)]).sum().round(decimals = 2)
gb_yoy
```

```python
# Plot the group by year values
gb_yoy.hvplot.bar(title='Trading Algorithm P/L by Year',yformatter='%.0f',xlabel='Year')
```

```python
# Group by month for a shot at seasonality
gb_mom = mom_df.groupby([(mom_df.index.month)]).sum().round(decimals = 2)
gb_mom
```

```python
gb_mom.hvplot.bar(title='Trading Algorithm P/L by Month - Seasonality',yformatter='%.0f',xlabel='Month')
```

```python
# Create df of entry exit holdings for plotting
holdings_df=qqq_df.loc[:,['Exit Portfolio Holding']]
holdings_df.tail()  
```

```python
port_df = holdings_df.groupby([(holdings_df.index.year)]).max().round(decimals = 2)
port_df
```

```python
port_df.hvplot.bar(yformatter='%.0f',xlabel='Year', title='Portfolio Balance by Year',ylabel='Account Balance')
```
